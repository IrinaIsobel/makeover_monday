```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vR8rt4hWcz9PR66hbDrmgLXpgfnnIi7vqhh00o9ZTweehedwK_6Xq3vP7almyhd2g/pub?output=csv')
df.head()
```

```python
fig, ax = plt.subplots(figsize = (13, 5))
ax = sns.histplot(data = df['Age at first pregnancy'],
                 edgecolor = 'white')
ax.set_xlabel('Age',
           fontsize = 13)
ax.set_xticks(np.arange(10, 35, 1))
ax.set_title('77% of women were teenagers and children when they had their first child')
ax.set_ylim(bottom = 0, top = 94);
plt.suptitle('Age at First Pregnancy', fontsize = 20)
ax.get_yaxis().set_visible(False)
ax.bar_label(ax.containers[0], label_type='edge', rotation = 90, padding = 3);
```
